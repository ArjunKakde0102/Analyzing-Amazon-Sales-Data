```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
#connecting to mysql 
engine = create_engine('mysql+pymysql://root:**************:3306/amazon')
```

```python
engine
```

```python
#import csv file from local machune
data = pd.read_csv("Amazon Sales Records (2).csv")
```

```python
data
```

```python
sales_data = data.drop(columns = ["Order Date","Ship Date"])
sales_data
```

```python
sales_data["Order Date"] = pd.to_datetime(data["Order Date"])
sales_data
```

```python
sales_data["Ship Date"] = pd.to_datetime(data["Ship Date"], format='mixed')
sales_data
```

```python
sales_data.to_sql(name = "sales", index =False, con = engine, if_exists = 'append')
```
